```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sn
import plotly.express as px
import json
import datetime
#import geopandas as gpd
import fiona

pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 1000)
```

```python
pluto_path = 'C:\\Users\\comp\\Documents\\New York\\nyc_mappluto_20v3_fgdb\\MapPLUTO20v3.gdb'

layers = fiona.listlayers(pluto_path)

pluto = fiona.open(pluto_path, layer = layers[0], mode = 'r')
```

```python
list_pluto = []
for rec in pluto:
    #print(rec)
    list_pluto.append([rec['properties']['CD'],
                      rec['properties']['LandUse'],
                      rec['properties']['Shape_Area']])
pluto.close()
```

```python
df_pluto = pd.DataFrame(list_pluto, columns = ['CD', 'LandUse', 'Shape_Area'])
df_pluto.head(100)
```

```python
df_pluto1 = df_pluto.groupby(['CD', 'LandUse'])['Shape_Area'].sum()
```

```python
df_pluto2 = pd.DataFrame(df_pluto1).reset_index()
df_pluto2['total'] = df_pluto2.groupby('CD')['Shape_Area'].transform(lambda x : x.sum(skipna = True))
df_pluto2['percent'] = df_pluto2['Shape_Area'] / df_pluto2['total']
df_pluto2
```

```python
df_pluto3 = df_pluto2.groupby(['CD', 'LandUse']).agg({'percent' : 'mean'}).unstack().fillna(value = 0.0)
df_pluto3.index = df_pluto3.index.astype(int)
df_pluto3.to_csv('pluto_land_use_percent.csv')
```

```python
from shapely.geometry import asShape

pluto = fiona.open(pluto_path, layer = layers[0], mode = 'r')
list_pluto_addr = []
for rec in pluto:
    #print(rec)
    s = asShape(rec['geometry'])
    p = s.representative_point()
    list_pluto_addr.append([rec['properties']['CD'],
                            rec['properties']['Borough'],
                            rec['properties']['Block'],
                            rec['properties']['Lot'],
                            rec['properties']['Address'],
                            [p.x, p.y]])
    
# apparently borough-block-lot is how addresses are looked up in NYC
# https://portal.311.nyc.gov/article/?kanumber=KA-01247

pluto.close()
```

```python
from address_parser import Parser

parser = Parser()
adr = parser.parse(rec['properties']['Address'])
adr.number.tnumber
adr.road.name
adr.road.suffix

```

```python
df_pluto_addr = pd.DataFrame(list_pluto_addr, columns = ['CD', 'Borough', 'Block', 'Lot', 'Address', 'Coords'])
df_pluto_addr.groupby('Borough').count()
```
